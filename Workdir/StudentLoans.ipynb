````python
import pandas as pd

data = pd.read_excel("PortfolioSummary.xls")

data.head()
````

````python
quarter_map = {
    "2013Q1": 2013,
    "2013Q2": 2013.25,
    "2013Q3": 2013.50,
    "2013Q4": 2013,
    "2014Q1": 2014,
    "2014Q2": 2014.25,
    "2014Q3": 2014.50,
    "2014Q4": 2014.75,
    "2015Q1": 2015,
    "2015Q2": 2015.25,
    "2015Q3": 2015.50,
    "2015Q4": 2015,
    "2016Q1": 2016,
    "2016Q2": 2016.25,
    "2016Q3": 2016.50,
    "2016Q4": 2016,
    "2017Q1": 2017,
    "2017Q2": 2017.25,
    "2017Q3": 2017.50,
    "2017Q4": 2017,
    "2018Q1": 2018,
    "2018Q2": 2018.25,
    "2018Q3": 2018.50,
    "2018Q4": 2018,
}
````

````python
import matplotlib.pyplot as plt

years = [quarter_map[j] if j in quarter_map else j for j in [i for i in data[data.keys()[0]][1:].keys() if (i not in quarter_map or "Q4" in i)]]
all_debts = [
    [data[j][i+1]
     for i in range(len(data[j][1:])) if (not type(data[data.keys()[0]].keys()[i+1]) is str) or "Q4" in data[data.keys()[0]].keys()[i+1]]
     for j in data.keys()]


plt.plot(
    years, all_debts[0], 
    #years, all_debts[1], 
    years, all_debts[2], 
    #years, all_debts[3], 
    years, all_debts[4], 
    #years, all_debts[5], 
    years, all_debts[6], 
    #years, all_debts[7]
)

plt.title("Student Loans Granted vs Time")
plt.xlabel("Fiscal year and/or quarter")
plt.ylabel("Outstanding Debt in Billion dollars")

plt.legend(("Direct Loans", "Federal Family Ed. Loans", "Perkins Loans", "Total"), bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()

plt.figure()

plt.plot(
    #years, all_debts[0], 
    years, all_debts[1], 
    #years, all_debts[2], 
    years, all_debts[3], 
    #years, all_debts[4], 
    years, all_debts[5], 
    #years, all_debts[6], 
    years, all_debts[7]
)

plt.title("Loan Recipients vs Time")
plt.xlabel("Fiscal year and/or quarter")
plt.ylabel("Number of loan recipients in millions")

plt.legend(("Direct Loans", "Federal Family Ed. Loans", "Perkins Loans", "Total"), bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
avg_loan_per_recipient = [i * 1000.0 / j for i, j in zip(all_debts[6], all_debts[7])]
plt.plot(
    years, [i * 1000.0 / j for i, j in zip(all_debts[0], all_debts[1])],
    years, [i * 1000.0 / j for i, j in zip(all_debts[2], all_debts[3])],
    years, [i * 1000.0 / j for i, j in zip(all_debts[4], all_debts[5])],
    years, [i * 1000.0 / j for i, j in zip(all_debts[6], all_debts[7])]
)

plt.title("Loan Amount per Recipient vs Time")
plt.xlabel("Fiscal year and/or quarter")
plt.ylabel("Dollars Loaned per recipient in dollars per recipient")

plt.legend(("Direct Loans", "Federal Family Ed. Loans", "Perkins Loans", "Total"), bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
````

# Income Data:

## Data Source: 
    https://www.census.gov/data/tables/time-series/demo/income-poverty/cps-finc/finc-07.html
    Data is from 2017

````python
income_data = pd.read_excel("finc07.xls", header=9)
income_keys = income_data.keys()
````

````python
income_keys
````

````python
labels = [i.replace(".", "").replace("$", "").replace("to", "-") if not "Under" in i else "2,500" for i in income_data[income_keys[0]]]

income_totals = [i * j for i, j in zip(income_data[income_keys[1]], income_data[income_keys[2]])]

ticks = [sum([float(j) for j in i.replace(',', '.').split(' - ')]) / 2.0
                     if not 'above' in i else 275 for i in labels]
````

````python
import matplotlib.pyplot as plt
````

````python
plt.figure(figsize=(80,40))
plt.bar(range(len(ticks)), income_totals)
plt.tick_params(labelsize=40)
plt.xticks(range(len(ticks)), labels, rotation=75)
plt.title("2017 U.S. Household Income Distribution", fontsize=90)
plt.xlabel("Income Group", fontsize=70)
plt.ylabel("Total Income in billions of dollars", fontsize=80)

plt.show()

#print("\n\n\n")

# plt.figure(figsize=(80,40))
# plt.bar(range(len(ticks)), income_data[income_keys[2]])
# plt.tick_params(labelsize=40)
# plt.xticks(range(len(ticks)), labels, rotation=75)
# plt.title("2017 U.S. Household Income Distribution", fontsize=90)
# plt.xlabel("Income Group", fontsize=70)
# plt.ylabel("Average Income in dollars", fontsize=80)

# plt.show()
# #print("\n\n\n")

plt.figure(figsize=(80,40))
plt.bar(range(len(ticks)), income_data[income_keys[1]])
plt.tick_params(labelsize=40)
plt.xticks(range(len(ticks)), labels, rotation=75)
plt.title("2017 U.S. Household Income Distribution", fontsize=90)
plt.xlabel("Income Group", fontsize=70)
plt.ylabel("Number of Households in thousands", fontsize=80)

plt.show()
````

````python
print(">100 total / all incomes:               ", sum([i for i in income_totals[-4:]]) / sum([i for i in income_totals]) *100)
print("percent of housholds making >100,000:   ", sum([i for i in income_data[income_keys[1]][-4:]]) / sum([i for i in income_data[income_keys[1]]]) *100)
print("Average household income:               ", sum([i for i in income_totals]) / sum([i for i in income_data[income_keys[1]]]))
````

# Total Income among Familes with incomes >100,000 / Total income of all households:

## 68.5% Of household income is held by familes of incomes >100,000

## That means 68.5% of the total of household incomes goes to  36.4%  of households

````python
import numpy as np

nums = income_data[income_keys[1]]
median = np.median(np.concatenate([np.full((nums[i]), ticks[i]) for i in range(len(ticks))]))
print("median income: $", median*1000)
print("really means median income is in group: $", labels[[i for i in range(len(ticks)) if ticks[i] == median][0]])

````
